```python
import numpy as np
np.random.seed(1234)
import scipy.io
import sys
import matplotlib.pyplot as plt
```

```python
plt.rcParams['xtick.labelsize'] = 20
plt.rcParams['ytick.labelsize'] = 20
plt.rcParams['axes.labelsize'] = 24
plt.rcParams['axes.titlesize'] = 20
plt.rcParams["savefig.bbox"] = 'tight'
plt.rcParams["savefig.pad_inches"] = 0.1
plt.rcParams['legend.fontsize'] = 16
```

```python
data_FEM = scipy.io.loadmat("../Data/CM2D2_FEM.mat")
obs_FEM = data_FEM["obs"].flatten()/1e3
rho_FEM_TE= data_FEM["rho_TE"].flatten()
phs_FEM_TE= data_FEM["phs_TE"].flatten()
rho_FEM_TM= data_FEM["rho_TM"].flatten()
phs_FEM_TM= data_FEM["phs_TM"].flatten()


H_c = np.loadtxt("../Data/CM2D2_10s.dat")
obs_c = H_c[:,0]
rho_TE = H_c[:,1]
rho_TE_er = H_c[:,2]
rho_TM = H_c[:,3]
rho_TM_er = H_c[:,4]
```

```python
data_TE_s= scipy.io.loadmat("../Data/CM2D2_TE_s.mat")
obs_TE_s    = data_TE_s["obs"].flatten()/1e3
rho_TE_s    = data_TE_s["rho_a"].flatten()
phs_TE_s    = data_TE_s["phs_a"].flatten()

data_TE_t= scipy.io.loadmat("../Data/CM2D2_TE_t.mat")
obs_TE_t    = data_TE_t["obs"].flatten()/1e3
rho_TE_t    = data_TE_t["rho_a"].flatten()
phs_TE_t    = data_TE_t["phs_a"].flatten()

data_TM_s= scipy.io.loadmat("../Data/CM2D2_TM_s.mat")
obs_TM_s    = data_TM_s["obs"].flatten()/1e3
rho_TM_s    = data_TM_s["rho_a"].flatten()
phs_TM_s    = data_TM_s["phs_a"].flatten()

data_TM_t= scipy.io.loadmat("../Data/CM2D2_TM_t.mat")
obs_TM_t    = data_TM_t["obs"].flatten()/1e3
rho_TM_t    = data_TM_t["rho_a"].flatten()
phs_TM_t    = data_TM_t["phs_a"].flatten()
```

```python
fig = plt.figure(figsize=(18,10))
ax = plt.subplot(2,2,1)
ax.plot(obs_FEM, rho_FEM_TE,'bp-',label="FEM")
ax.plot(obs_TE_t, rho_TE_t,'gp-',label="TFM")
ax.plot(obs_TE_s, rho_TE_s,'rp-',label="SFM")
# ax.plot(obs_TE_c, rho_TE_c,'kD', markersize=10,label="COMMEMI")
ax.errorbar(obs_c, rho_TE,yerr=rho_TE_er,capsize=3,color='black',marker='d',linestyle='',label="COMMEMI")

# ax.set_yscale("log")
ax.set_ylabel(r'$\rho_a$($\Omega$m)')
ax.set_title('(a)',loc='center')#,fontdict={'fontSize':20})
ax.legend(loc='lower left',fontsize=12.5)
# ax.ticklabel_format(style='plain',axis='y')

ax = plt.subplot(2,2,3)
ax.plot(obs_FEM, phs_FEM_TE,'bp-',label="FEM")
ax.plot(obs_TE_t, phs_TE_t,'gp-',label="TFA")
ax.plot(obs_TE_s, phs_TE_s,'rp-',label="SFA")
ax.set_ylabel(r'$\phi$(degree)')
ax.set_xlabel('distance(km)')
ax.set_title('(c)',loc='center')#,fontdict={'fontSize':20})


ax = plt.subplot(2,2,2)
ax.plot(obs_FEM, rho_FEM_TM,'bp-',label="FEM")
ax.plot(obs_TM_t, rho_TM_t,'gp-',label="TFM")
ax.plot(obs_TM_s, rho_TM_s,'rp-',label="SFM")

# ax.plot(obs_c, rho_c,'kD', markersize=10,label="COMMEMI")
ax.errorbar(obs_c, rho_TM,yerr=rho_TM_er,capsize=3,color='black',marker='d',linestyle='',label="COMMEMI")

# ax.plot(obs_c, rho_c,color='purple',marker, markersize=10,label="COMMEMI")
# ax.set_yscale("log")
# ax.set_ylabel(r'$\rho_a$($\Omega$m)')
ax.set_title('(b)',loc='center')#,fontdict={'fontSize':20})
ax.legend(loc='lower left',fontsize=12.5)
# ax.ticklabel_format(style='plain',axis='y')

ax = plt.subplot(2,2,4)
ax.plot(obs_FEM, phs_FEM_TM,'bp-',label="FEM")
ax.plot(obs_TM_t, phs_TM_t,'gp-',label="TFA")
ax.plot(obs_TM_s, phs_TM_s,'rp-',label="SFA")
# ax.set_ylabel(r'$\phi$(degree)')
ax.set_xlabel('distance(km)')
ax.set_title('(d)',loc='center')#,fontdict={'fontSize':20})


# plt.subplots_adjust(wspace =0.25)#

# plt.savefig("./imag/TE_TM_2_2.jpg",dpi=300,bbox_inches='tight',pad_inches=0.05)
plt.show()
```
